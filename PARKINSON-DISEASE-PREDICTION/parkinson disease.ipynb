```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
df=pd.read_csv('Parkinsson disease.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
data={'MDVP:Fo(Hz)':'col1','MDVP:Fhi(Hz)':'col2','MDVP:Flo(Hz)':'col3','MDVP:Jitter(%)':'col4','MDVP:Jitter(Abs)':'col5',
     'MDVP:RAP':'col6','MDVP:PPQ':'col7','Jitter:DDP':'col8','MDVP:Shimmer':'col9'}
```

```python
df.rename(columns=data,inplace=True)
```

```python
data1={'MDVP:Shimmer(dB)':'col10', 'Shimmer:APQ3':'col111','Shimmer:APQ5':'col122','MDVP:APQ':'col13','Shimmer:DDA':'col14',
      'NHR':'col15','HNR':'col16','RPDE':'col17','DFA':'col18','spread1':'col19','spread2':'col20','D2':'col21','PPE':'col22'}
```

```python
df.rename(columns=data1,inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['status'].value_counts()
```

```python
df.groupby('status').mean()
```

```python
x=df.drop(columns=['name','status'],axis=1)
y=df['status']
```

```python
x
```

```python
x.columns
```

```python
y.head()
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(x)
```

```python
s_df=scaler.transform(x)
```

```python
print(s_df)
```

```python
X=s_df
Y=df['status']
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=y,random_state=2)
```

```python
model=SVC(kernel='linear')
model.fit(X_train,Y_train)
```

```python
train_shape=[x.shape,X_train.shape,Y_train.shape]
test_shape=[X_test.shape,Y_test.shape]
```

```python
print(train_shape)
print(test_shape)
```

```python
y_pred=model.predict(X_train)
accuracy=accuracy_score(Y_train,y_pred)
```

```python
print(accuracy)
```

```python
y1_pred=model.predict(X_test)
accuracy=accuracy_score(y1_pred,Y_test)
```

```python
print(accuracy)
```

```python
input_data=(119.992,157.302,74.997,0.00784,0.00007,0.0037,0.00554,0.01109,0.04374,0.426,0.02182,0.0313,0.02971,0.06545,0.02211,21.033,0.414783,0.815285,-4.813031,0.266482,2.301442,0.284654)
in_array=np.asarray(input_data)
in_reshape=in_array.reshape(1,-1)
scale_df=scaler.transform(in_reshape)
prediction=model.predict(scale_df)
print(prediction)

if (prediction[0]==0):
    print('the person does not have diabetic disease')
else:
    print('the person have diabetic disease')
```
